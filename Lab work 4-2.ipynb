#### Ознайомитися із API Національного банку

1. Підключитися до API.


url = https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json

```
import requests as r

url = 'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json'
responds = r.get(url)
print(responds.status_code)
```

2. Отримати курс долара за останній рік.

```
from pprint import pprint

params = {'start':'20210101', 'end': '20211231', 'valcode': 'usd'}
url1 = 'https://bank.gov.ua/NBU_Exchange/exchange_site?sort=exchangedate&order=desc&json'

responds = r.get(url1, params=params)

exc_rate = responds.json()
pprint(exc_rate)
```

3. Вивести середнє значення та відхилення курсу за кожний місяць.

```
import datetime
import statistics

dict_curr = {}
for i in exc_rate:
    month = datetime.datetime.strptime(i['exchangedate'], "%d.%m.%Y").month
    if month not in dict_curr: 
        dict_curr[month]=[i['rate']]
    else:
        dict_curr[month].append(i['rate'])   
        

pprint(dict_curr)


```

```
result_std = {i: [statistics.mean(dict_curr[i]),statistics.stdev(dict_curr[i])] for i in dict_curr}
pprint(result_std)
```

4. Дану інформацію записати у файл за допомогою pickle.

```
import pickle
with open('my_result.picke', 'wb') as file: 
    pickle.dump(result_std, file)
```

### Завдання 2:

Потрібно проаналізувати всі товари на сайті:

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```
from bs4 import BeautifulSoup 

dict_site = {}
url_agro = 'https://smallpacking.agrosem.ua/shop/'
responds_new = r.get(url_agro)

soup = BeautifulSoup(responds_new.text, 'lxml')
elements = soup.find_all('div', class_="product-from-category-container")
page_number = soup.find('ul', class_="page-numbers").find_all('a', class_="page-numbers")[-2].text 
for i in range(1, int(page_number)+1): 
    a = 'https://smallpacking.agrosem.ua/shop/page/' + str(i)  + '/'
    responds_all = r.get(a)
    soup = BeautifulSoup(responds_all.text, 'lxml')
    elements = soup.find_all('div', class_="product-from-category-container")
    for element in elements:
        name = element.find('span', class_="pruduct-name-bottom").text
        price = element.find('span', class_="regular-price").text[:-4].replace(' ', '')
        link = element.find('a').get('href')
        responds_new2 = r.get(link)
        soup_new = BeautifulSoup(responds_new2.text, 'lxml')
        sku = soup_new.find('span', class_="sku").text[5:]
        try: 
            weight = soup_new.find('tr', class_="woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_vaga").find('td', class_="woocommerce-product-attributes-item__value").text.rstrip()
        except:
            weight = None
        if len(price) == 0: 
            price = 0
        price_ = float(price)
        dict_site[sku] = [name, price_, weight] 
pprint(dict_site)
```
