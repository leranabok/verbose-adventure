Напишіть функції, що:

* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)] 

 
```

```
print(list_with_random_numbers[:10])
```

```
sorted(list_with_random_numbers[:10])
```

```
def max_value(a): 
    max_ = a[0]
    for i in a:
        if i > max_: 
            max_ = i
    return max_
        
```

```
max_value(list_with_random_numbers[:10])
```

```
def min_value(a): 
    min_ = a[0]
    for i in a:
        if i < min_: 
            min_ = i
    return min_
```

```
min_value(list_with_random_numbers[:10])
```

```
def average_value(a):
    return sum(a)/len(a)
```

```
average_value(list_with_random_numbers[:10])
```

```
def mediana_value(a): 
    b = int((len(a)/2)-1)
    return ((a[b]+a[b+1])/2)
```

```
import math 
def std(a):
    new_list = [(x-average_value(a))**2 for x in a]
    return math.sqrt(sum(new_list)/(len(a)-1))
```
